**Zipf's law** describes a relationship between the frequencies and ranks of words in natural languages.  The "frequency" being the number of times a word appears in a body of work.  The "rank" being its position in a list of words sorted by frequency.

$$
f=cr^{-s}
$$
where s and c are parameters that depend on the language and text.  If you take the log of both sides and plot log(f) versus log(r) you should get a straight line with slope -s and intercept log c

$$
\log f = \log c-s \log r
$$

```python
%matplotlib inline
import string
from collections import defaultdict
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import thinkplot, thinkstats

def remove_punctuation(s):
    exclude = set(string.punctuation)
    exclude.discard('\\')
    exclude.discard('-')
    exclude.add(u'\u201c') # double quotation mark
    exclude.add(u'\u201d') #
    exclude.add(u'\u2019') # '
    exclude.add(u'\u2014') # m-dash
    
    return ''.join([ch.lower() for ch in s if ch not in exclude])
    
    
def make_word_hist(txt='100years.txt'):
    import codecs
    
    words = defaultdict(int)
    f = codecs.open(txt, encoding='utf-8')
    for line in f:
        line = remove_punctuation(line)
        line = line.split()
        for w in line:
            words[w] += 1
    freqs = words.values()
    freqs.sort(reverse=True)
    return freqs


    
freqs = make_word_hist()


def find_s(freqs):
    """
    logf = logc - s*logr
    
    but for now assume logc is small
    """
    logf = np.log10(freqs, dtype=np.float64)
    logr = np.log10([i + 1 for i in xrange(len(freqs))])
    logr = logr.reshape((-1,1))
    
    assert not np.any(np.isnan(logf))
    assert not np.any(np.isinf(logf))
    assert not np.any(np.isnan(logr)) 
    assert not np.any(np.isinf(logr))
    assert len(logf) == len(logr)
    
    from sklearn import linear_model
    regr = linear_model.LinearRegression()
    regr.fit(logr, logf)
    return -regr.coef_[0], regr.intercept_

def plot_s(freqs, s=None, ax=None):
    if s == None:
        s, logc = find_s(freqs)
    rs = np.linspace(1, len(freqs), len(freqs))
    logr = np.log10(rs)
    logf = logc - s*logr
    
    if ax is None:
        fig, ax = plt.subplots()
    ax.plot(logr, logf, label='s = %g' % s)
    return s, logc
    
    
def make_words_cdf(freqs, ax=None):
    """
    freqs: sorted list of frequencies
    """
    cumulative = np.cumsum(freqs)
    logcum = np.log10(freqs)
    logr = np.log10(range(len(freqs)))
    
    if ax is None:
        fig, ax = plt.subplots()
    ax.plot(logr, logcum)


fig, ax = plt.subplots()
ax.set_xlabel('log word rank')
ax.set_ylabel('log word frequency')
ax.set_title('Word Rank and Frequencies in' +
             '100 Years of Solitude (Log scale)')
# ax.set_xscale('log')
# ax.set_yscale('log')

make_words_cdf(freqs,ax=ax)
s, c = plot_s(freqs, ax=ax)
print "slope:", s
print "intercept", c
plt.legend(loc='best')
plt.show()
```

##Exponential Distribution
$$
CDF(x) = 1 - e^{-\lambda x}
$$
<br>
$$
CCDF = y = 1 - CDF(x) \sim e^{-\lambda x}
$$
<br>
$$
\log y \sim -\lambda x
$$

$\lambda$ can be got from negative slope by plotting on a logy scale

## exercise 5.3


```python
import thinkplot
def plot_ccdf_logy(cdf, ax=None, label=''):
    values, probs = zip(*cdf.Items())
    probs = np.cumsum(probs)
    ys = [1.0-p for p in probs]
    if ax is None:
        fig, ax = plt.subplots()
    ax.plot(values, ys, label=label)
    ax.set_yscale('log')
    return ax

n = 100
res = np.random.exponential(scale=4, size=n)
cdf = thinkstats.Cdf(res)
ax = plot_ccdf_logy(cdf)
ax.set_xlabel('values')
ax.set_ylabel('probabilites of being less than value')
ax.set_title('CCDF of exponential function (logscaled)')
thinkplot.Show()
```

##Pareto Distribution

$$
CDF(x) = 1 - \Big(\frac{x}{x_m}\Big)^{-\alpha}
$$
where $x_m$ and $\alpha$ determine the location and shape of the distribution; $x_m$ is the minimum possible quantity

####Properties:
* **Long Tail**: many small values and few very large ones.
* **80/20 rule**: the large values in a Pareto are so larget that they make up a disproportionate share of the total value.
* **scale free**: there is no similar range that contains the bulk of the distribution.

on a log-log scale, the CCDF looks like a straight line.

$$
y = 1-CDF(x) \sim \Big(\frac{x}{x_m}\Big)^{-\alpha}
$$
<br>
$$
\log y \sim -\alpha(\log x - \log x_m)
$$

```python
import thinkplot
def plot_ccdf_log_log(cdf, ax=None, label=''):
    values, probs = zip(*cdf.Items())
    probs = np.cumsum(probs)
    ys = [1.0-p for p in probs]
    if ax is None:
        fig, ax = plt.subplots()
    ax.plot(values, ys, label=label)
    ax.set_yscale('log')
    ax.set_xscale('log')
    return ax

n = 100000
res = np.random.pareto(.8, size=n)
cdf = thinkstats.Cdf(res)

ax = plot_ccdf_log_log(cdf)
ax.set_xlabel('values')
ax.set_ylabel('probabilites of being less than value')
ax.set_title('CCDF of pareto variated pseudo-random values (logscaled)')
plt.show()
```

The curve straightens out on the log-log scale as you increase the amount of values.  For fewer values, the curve resembles that of an exponential.

##Barabasi and Albert
"Emergence of Scaling in Random Networks"
* measure degree of each node and compute P(k), probability that a vertex has degree k
* plot P(k) versus k on a log-log scale
* tail fits straight line, so obeys **power law**; that is, as k gets large, p(k) is asymptotic to $k^{-\gamma}$.  
* distribution of degrees is 'scale-free'.  Note, sometimes power law distributions are called **scaling distributions**, because they are invariant under a change in scale.  If you change the units, the slope parameter, $\gamma$ does not change.

diffent from E-R model and W-S model because:
* **growth**: instead of starting with a fixed number of vertices, B+A start witha small graph and add vertices gradually.
* **Preferential attachment**: When a new edge is created, it is more likely to connect to a vertex that already has a large number of edges.  

In ER and WS models, number of vertices with high number of attachments decreases decreasees exponentially with k, so highly connected vertices are practically absent.

##Exercise 5.6

[paper link](http://www.barabasilab.com/pubs/CCNR-ALB_Publications/199910-15_Science-Emergence/199910-15_Science-Emergence.pdf).  the genrative model algorithm goes as follows:
1.  start with a small number $m_0$ of vertices.
2.  at every time step add a new vertex with $m(\leq m_0)$ edges.  probability $\Pi$ that new vertex will be connected to vertex $i$ depends on the connectivity $k_i$ of that vertex.  So $\Pi(k_i) = k_i/ \sum_j k_j$, where j denotes other vertices

after t time steps:
* random network with $t + m_0$ vertices and $mt$ edges
* probability that a vertex has k edges follows a pwer law with an exponent $\gamma_{model} = 2.9 \pm 0.1$.

```python
from SmallWorldGraph import *
from Graph import *
from GraphWorld import ShowGraph

####1 
class BandAGraph(SmallWorldGraph):
    
    def __init__(self, m_0=5, k=2):
        """
        m_0 unconnectect vertices are created.
        A is inital attractiveness of each node.
        
        v.k stores connectivity of a single vertex
        self.sum_k stores sum of all v.k
        """
        Graph.__init__(self)
        self.names = self.name_generator()
        self.random_vertice = self.vertice_generator()
        for i in range(m_0):
            v = Vertex(self.names.next())
            v.k = k
            self.add_vertex(v)
        self.add_regular_edges(k=k)
        self.sum_k = m_0 * k   
    
    def vertice_generator(self):
        vertices = self.vertices()
        while True:
            np.random.shuffle(vertices)
            for v in vertices:
                yield v
        
        
    def name_generator(self):
        i = 0
        while True:
            i += 1
            yield i
    
    def make_connection(self, v):
        while True:
            w = self.random_vertice.next()
            if w != v: break
        threshold = w.k / float(self.sum_k - w.k)
        if np.random.random() < threshold:
            self.add_edge(Edge(v, w))
            v.k += 1
            w.k += 1
            self.sum_k += 2
            
    def timestep(self, min_conns=1, v=None):
        if v==None:
            v = Vertex(self.names.next())
        v.k = 0
        self.add_vertex(v)
        
        while v.k < min_conns:
            for i in xrange(len(self.vertices())):
                self.make_connection(v)
        
    def n_timesteps(self, n, min_conns=1, ret=False):
        for i in range(n):
            if i % 5000 == 0:
                print '%d of %d timesteps: %g' % (i, n, etime())
            self.timestep(min_conns=min_conns)
        if ret:
            key=lambda v: v.k
            return sorted(self.vertices(), key=key)
            
# implement graph with 150000 vertices and replicate figure2
ba = BandAGraph(m_0=6,k=1)
ba.n_timesteps(2)
ShowGraph(ba)
print ba.is_connected()
```

```python
####2
# make largest possible WS graph
def _rewire(self, p=0.01):
    es = list(self.edges())
    random.shuffle(es)
    vs = self.vertices()

    for e in es:
        # if this edge is chosen, remove it...
        if random.random() > p: continue
        v, w = e
        self.remove_edge(e)
        v.k -= 1
        w.k -= 1

        # then generate a new edge that connects v to another vertex
        while True:
            w = random.choice(vs)
            if v is not w and not self.has_edge(v, w): break

        self.add_edge(Edge(v, w))
        v.k += 1
        w.k += 1
BandAGraph.rewire = _rewire

# plot p vs k for this graph
fix, ax = plt.subplots(nrows=3)
for p in [.01, .1, .7]:
    ws = BandAGraph(m_0=100000, k=10)
    ws.rewire(p=p)
    cdf = thinkstats.Cdf([v.k for v in ws.vertices()])
    plot_ccdf_log_log(cdf, ax=ax[0], label='p = %g' % p)
    plot_ccdf_logy(cdf, ax=ax[1], label='p = %g' % p)
    plot_ccdf_linear(cdf, ax=ax[2], label='p = %g' % p)
ax[0].set_title('Plotting connectivity cdf for Watts/Strogats\n' +
             'Graph with 100000 nodes')
for a in ax:
    a.set_ylabel('P(k)')
    a.set_xlabel('k')
    a.legend(loc='best')
fig.tight_layout()
plt.show()
```

[this is a good paper on this topic](http://www.hpl.hp.com/research/idl/papers/ranking/ranking.html)
all describe systems where large events are rare and scale free.  

* zipf - rank compared to size or frequency
* pareto - how many people have an income greater than x.  Given in terms of CDF
* power law distribution - tells us the amount of people whose income actually is x.  pdf and derivative of pareto law.

zipf and pareto are obviously very similar graphs but perhaps have different slopes, because they are basically different scaling on the x axis. saying the "rth largest city has n inhabitants" is very simalar to saying "r cities have n or more inhabitants" excpt that the axes are flipped.

##Explanatory Models
system of analogy.
1. in a system, S, we see something observable, O, that warrents explanation.
2. We construct a model, M that is analogous to the system.  Corresponence between elements of model and the elements of the system.
3. By simulation or derivation we show that the model M exhibits a behavior, B, similar to O
4.  We conclude that S exhibits O because S is similar to M, M exhibits B, and B is similar to O.  

---
* **model** ------derivation----> **Behavior**
*     |                                |
*     |abstraction                     |analogy
*     |                                |
* **system** -----observation---> **Observable**

WW model and BA model both exhibit small world behavior, bu they offer different explanations.

* WS model says social networks are small because they include both strongly connected clusters and weak ties that connect clusters.
* BA model suggest that social networks are small because of hubs.
